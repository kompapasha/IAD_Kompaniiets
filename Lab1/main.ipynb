```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
movie = ['MovieID', 'Title', 'Genres']
data_movies = pd.read_table('ml-1m/movies.dat',sep='::',engine='python',encoding='latin-1',names=movie)
data_movies
```

```python
rating = ['UserID','MovieID','Rating','Timestamp']
data_ratings = pd.read_table('ml-1m/ratings.dat',sep='::',engine='python',encoding='latin-1',names=rating)
data_ratings
```

```python
user = ['UserID','Gender','Age','Occupation','Zip-code']
data_users = pd.read_table('ml-1m/users.dat',sep='::',engine='python',encoding='latin-1',names=user)
data_users
```

```python
mergeone = pd.merge(data_users ,data_ratings, how='left', left_on='UserID', right_on='UserID')
mergeall = pd.merge(mergeone, data_movies, how='left', left_on='MovieID', right_on='MovieID')
mergeall
```

```python


ordered = mergeall.sort_values(by='MovieID', ascending=False)
ordered
```

```python
rating_gender = mergeall.pivot_table('Rating', index = 'Title', columns = 'Gender', aggfunc = 'sum')
rating_M = rating_gender.filter(items = 'M')
top_for_M = rating_M.sort_values(by = 'M', ascending = False).head(10)
top_for_M
```
