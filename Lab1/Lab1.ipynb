```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
movie = ['idmovie','name','description']
data_movie = pd.read_table('ml-1m/movies.dat',sep='::',engine='python',encoding='latin-1',names=movie)
data_movie
```

```python
ratings = ['iduser','idmovie','rating','timestamp']
data_rating = pd.read_table('ml-1m/ratings.dat',sep='::',engine='python',encoding='latin-1',names=ratings)
data_rating
```

```python
users = ['iduser','gender','age','occupation','zip_code']
data_users = pd.read_table('ml-1m/users.dat',sep='::',engine='python',encoding='latin-1',names=users)
data_users
```

```python
mergeone = pd.merge(data_users ,data_rating, how='left', left_on='iduser', right_on='iduser')
mergeall = pd.merge(mergeone, data_movie, how='left', left_on='idmovie', right_on='idmovie')
mergeall
```

```python
ordered = mergeall.sort_values(by='iduser')
ordered
```
