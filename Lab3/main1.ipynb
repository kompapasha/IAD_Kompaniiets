```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Life Expectancy Data.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
missing_data = [ 'Life expectancy ', 'Adult Mortality', 'Alcohol', 'Hepatitis B',
' BMI ',  'Polio', 'Total expenditure', 'Diphtheria ',  'GDP', 'Population',
' thinness  1-19 years', ' thinness 5-9 years','Income composition of resources',
'Schooling']

for i in missing_data:
    df.loc[df.loc[:,i].isnull(),i] = df.loc[:,i].median()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
column_num = df.drop(['Country','Year','Status'], axis = 1)
plt.figure(figsize=(12,60))

i = 1
for name in column_num:
    plt.subplot(10,2,i)
    sns.boxplot(column_num, y = name)
    i += 1
```

```python
plt.figure(figsize=(12,60))

i = 1
for name in column_num:
    plt.subplot(10,2,i)
    sns.histplot(df[name], bins= 25)
    sns.rugplot(df[name], height=-0.02, clip_on=False)
    i += 1
```

```python
pd.options.display.float_format = '{:.2f}'.format
variance = df.var()
variance
```
